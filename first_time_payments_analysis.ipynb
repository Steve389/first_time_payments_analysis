## 1. Correlation between user ID and card number.

### In a perfect world, if these are all first payments, to each unique user ID should correspond a unique card number. However, this isn't a perfect world, so the question we need to ask ourselves would be: are there any users that used the same card? Or are there cards that were used for a suspicious amount of transactions? This could be a signal of credit card theft and/or creation of many fake accounts.

Let's start with a quick overview of the card numbers and the user_ids which used them. 
After that, we'll show the card numbers that were used more than 4 times.

```python
fraud.groupby('card_bin')['user_id'].value_counts().head(50)
```

```python
cards_used_more_than_5 = fraud.card_bin.value_counts().loc[lambda x : x>4]

cards_used_more_than_5
```

These results show that several cards correspond to more than one user and that 1028 cards were used more than 5 times.
One useful development would be to set a filter that detects cards that were used more than a certain threshold even though they are linked to different users.

However, the fact that the same credit card is used more than once doesn't necessarily point to fraudulent action; let's consider this scenario: 4 people living in the same household have separate accounts but share and use the same credit card. In that situation, the fraudulent motive seems less likely. 
How to spot these false positives? 

- We could use the longitude and latitude data to see if different user IDs have placed orders with the same credit cards from the same location. However, we don't always order a bolt from the same location so what about the situations in which the users are in a different location than their home? And if we do find matching locations, how do we know that it's proof of a shared household?
- A better approach would be to identify if there are devices that share the same IP address ( WiFi router). Unfortunately, that is something that we can't do with this dataset.

## 4. GPS discrepancies. 

### Discrepancies between the actual distance (between the coordinates) and the ride_distance values. 
### In the dataset, there are the pickup coordinates, destination coordinates and the total distance of the ride. But what if users are manipulating the GPS data in order to make their trip appear shorter on the app (or longer in the case of drivers)? Or what if passengers are cancelling trips in the middle of a ride to trick the app and pay less? Calculating the distance between the coordinates and comparing it to the distance reported on the dataset might be a useful strategy to detect unsual patterns.

Let's first create a new dataset that includes only the coordinates data and then get rid of any null values.

```python
fraud_coordinates = fraud[['lat','lng','real_destination_lat','real_destination_lng','ride_distance']]
fraud_coordinates
```

```python
print(fraud_coordinates.columns[fraud_coordinates.isnull().any()])
```

```python
fraud_coordinates = fraud_coordinates.dropna(subset=['real_destination_lat', 'real_destination_lng'])
```

Let's create a function that will calculate the distance between the coordinates and append the results to a new column 'calculated_distances'.

```python
def distance_calculator(df,col1, col2,col3, col4):
   
    distances = []    
    
    for i in range(len(col1)):
        
        coord1 = (col1.values[i],col2.values[i])

        coord2 = (col3.values[i], col4.values[i])
       
        distances.append(geopy.distance.geodesic(coord1, coord2).m)
            
    df['calculated_distance']= distances
            

        

```

```python
distance_calculator(fraud_coordinates,fraud_coordinates['lat'],fraud_coordinates['lng'],fraud_coordinates['real_destination_lat'],fraud_coordinates['real_destination_lng'])
```

```python
fraud_coordinates
```

Next, let's create another function that will flag cases in which the difference between 'ride_distance' and 'calculated_distance' is higher than 5000 meters.

```python
def flag_raiser(df,col1,col2):
    
    flags=[]
    
    for i in range(len(col1)):
        if (col1.values[i]- col2.values[i]> 5000) | (col2.values[i]- col1.values[i]> 5000):
            flags.append('yes')
        else:
            flags.append('no')
                
    df['to_be_flagged']= flags

```

```python
flag_raiser(fraud_coordinates,fraud_coordinates['ride_distance'], fraud_coordinates['calculated_distance'])
```

```python
fraud_coordinates
```

Now we have a new column that shows us when there are these kind of GPS discrepancies and flags them.

## 5. Correlation between devices used and number of failed payments

### Is there a correlation between fraudulent activity and devices used? 

Let's start with a comparison between iOS and android devices.

```python
conditions = [
    (fraud['device_os_version'].str.contains('iOS')),
    (fraud['device_os_version'].str.contains('iOS')== False)
]

values = ['iOS','android']

fraud['OS']= np.select(conditions, values)
```

```python
sns.countplot(x= 'OS',data= fraud)


plt.xlabel('Operating System', fontsize=12)
plt.ylabel('failed_payments', fontsize=12)
plt.title('Distribution of Operating Systems', fontsize=18)
plt.show()
```

Overall, android devices account for the majority of failed payments. One thing we should keep in mind though, is that android OS is supported by several devices, while iOS is only supported by one device: the iPhone. This means that the iPhone transactions alone equal 3/4 of all android models transactions.

What if we shift our focus to specific devices and see which device models were used in more than 500 failed transactions?

```python
fraud_device =fraud.device_name.value_counts().loc[lambda x : x>500]

sns.lineplot(data=fraud_device)


plt.ylabel('failed payments', fontsize=12)
plt.xticks(rotation=90, fontsize=12)

plt.title('Number of failed payments per device', fontsize=18)

plt.show()
```

It looks like iPhone devices have been used the most. To validate this hypothesis, a useful strategy would be to cross reference the device models with credit card information and (if present in a different dataset) IP addresses and phone numbers of the users.

This could highlight interesting patterns:

- In step 1, we found out that many users share the same credit card number and that more than 1000 cards were used in more than 5 first payments. What if many of these users share the same device? This might be the same person that is creating many fake accounts using one device and one (probably stolen) card.

- If the same number and IP address matches more users, this might the same fraudster using the same device and number to create many fake accounts

- If the same IP address matches different numbers and users, these might be different fraudsters using different phone numbers but the same device.



Without IP address and telephone numbers data, we can't dig deeper in this analysis. One thing we can do, is select from the fraud dataset only the payments that were made with the same card at least 5 times. We have this information in the series we saved as 'cards_used_more_than_5' in step 1. 

```python
cards_used_more_than_5 = cards_used_more_than_5.to_frame().reset_index().rename(columns = {'index': 'card_bin','card_bin':'failed_transactions'}, inplace = False)

fraud_more_than_5 = fraud[fraud['card_bin'].isin(cards_used_more_than_5.card_bin.unique())]
```

Now that we have a dataset with only the card numbers from the series 'cards_used_more_than_5', let's see if also in this case, the highest number of failed transactions happened on IOS devices.

```python
fraud_more_than_5_device = fraud_more_than_5.device_name.value_counts().loc[lambda x : x>500]

sns.lineplot(data=fraud_more_than_5_device)

plt.xticks(rotation =90, fontsize=12)
plt.ylabel('failed payments', fontsize=12)
plt.title('Failed payments per device (Across card numbers that were used more than 5 times) ', fontsize=18)

plt.show()
```

As in the previous graph, the top devices are iPhones.
What if we restrict the selection to just the card numbers that were used more than 500 times?

```python
cards_used_more_than_500 = fraud_more_than_5.card_bin.value_counts().loc[lambda x : x>500]
cards_used_more_than_500 = cards_used_more_than_500.to_frame().reset_index().rename(columns={'index':'card_bin','card_bin':'failed_transactions'}, inplace = False)
fraud_more_than_500 = fraud_more_than_5[fraud_more_than_5['card_bin'].isin(cards_used_more_than_500.card_bin.unique())]                                                                        
                                            
```

```python
fraud_more_than_500_device = fraud_more_than_5.device_name.value_counts().loc[lambda x : x>500]

sns.lineplot(data=fraud_more_than_500_device)

plt.xticks(rotation =90, fontsize=12)
plt.ylabel('failed payments', fontsize=12)
plt.title('Failed payments per device (Across card numbers that were used more than 500 times) ', fontsize=18)

plt.show()
```

Even in this case, things don't change. If by cross matching IP address, phone numbers and credit card info we can verify that the same device was used to create multiple users, we might be on to something.

If that were the case, more questions would arise:

- Why were iPhone devices mostly used? Is it just because of their widespread adoption or is there a reason(s) to why a fraudster might prefer this device over another one?

- Is this iPhone domination misleading? What if the fraudsters using apple devices are not the smartest hackers around and that's why it was easy for us to identify them?

- What if the fact that many different android devices were being used for fewer transactions is actually a hidden signal of intelligent fraudsters that know their actions cannot be connected to one single device?

## 6. Distinction between 'real' cancellations and 'fake' ones.

### This dataset doesn't contain information  about cancellations,  but it might be useful to compare the information found in step 1 with cancellation data. Trip cancellation doesn't always mean fraud, it also might be a 'normal' customer who genuinely changed his mind or his plans. In light of this, it would be key to distinguish between 'good' and 'bad' cancellations.
### If the same users who appear to share the same credit card info appear to have also high cancellation rates, then this might be a pattern. For example, systemic cancellations are frequently in the case of so called friendly fraud, when the driver tells the passenger to cancel the trip and pay him directly, in this way excluding the riding service from the process. In other cases, strategic cancellations are employed as a strategy to disrupt businesses.
